### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', 
                    sep='\t', 
                    index_col='order_id')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
# Changing item_price to float
price_c = [float(value[1:-1]) for value in chipo.item_price]
# price_c is a list, set item_price series = to price_c
chipo.item_price = price_c
chipo.item_price.dtype
```

```python
chipo[chipo['item_price'] > 10]['item_price'].count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.count()
```

```python
# Drop duplicates first 
c_item_name_no_dupes = chipo.drop_duplicates(['item_name'])
c_item_name_no_dupes.sort_values(['item_price'])
```

### Step 6. Sort by the name of the item

```python
c_item_name_no_dupes.sort_values(['item_name'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive = chipo.sort_values(['item_price'], ascending=False).head(1)
most_expensive['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].count()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo_soda = chipo[chipo['item_name'] == 'Canned Soda']
chipo_soda[chipo_soda['quantity'] > 1].count()[1]
```

```python
chipo_soda = chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)].count()[1]
chipo_soda
```
